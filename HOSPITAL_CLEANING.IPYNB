```python
import matplotlib.pyplot as plt
import seaborn as sns

age_filtered =df[df['Age'] <=100]

sns.histplot(age_filtered['Age'],bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("count")
plt.show()
```

```python
sns.countplot (
    y="Treatment_Type",
    data=df
)
plt.title("Treatment Type Distribution")
plt.xlabel("count")
plt.ylabel("Treament Type")
plt.show()
```

```python
sns.countplot(
    y="Doctor_Name",
    data=df
)
plt.title("Patients Per Doctor")
plt.xlabel("Number of patients")
plt.ylabel("Doctor Name")
plt.show()
```

```python
sns.countplot(
    data=df,
    x='Hospital_Ward'
)

plt.title("Patient Count by Hospital Ward")
plt.xlabel("Hospital Ward")
plt.ylabel("Number of Patients")
plt.show()
```
